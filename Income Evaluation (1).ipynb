```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df[' workclass'] = le.fit_transform(df[' workclass'])
df[' education'] = le.fit_transform(df[' education'])
df[' marital-status'] = le.fit_transform(df[' marital-status'])
df[' occupation'] = le.fit_transform(df[' occupation'])
df[' relationship'] = le.fit_transform(df[' relationship'])
df[' race'] = le.fit_transform(df[' race'])
df[' sex'] = le.fit_transform(df[' sex'])
df[' native-country'] = le.fit_transform(df[' native-country'])
df[' income'] = le.fit_transform(df[' income'])
```

```python
#dropping the fnlwgt column because of its negligible correlation with income
df.drop(' fnlwgt', axis=1, inplace= True)
```

```python
df.head()
```

```python
x = df.iloc[:,0:13]
y = df.iloc[:,13]
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
ss= StandardScaler()
ss.fit(x)
x = ss.transform(x)
```

```python
x
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

# KNN Classifier

```python
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn import metrics
from sklearn.metrics import classification_report
knn = KNeighborsClassifier(n_neighbors=25, metric='euclidean')
knn.fit(x_train, y_train)
```

```python
y_pred = knn.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

# Decision Tree Classifier as Entropy

```python
from sklearn.tree import DecisionTreeClassifier as dtree
ct = dtree(
           criterion="entropy",    # Alternative 'gini'
           max_depth=None       # Alternative, specify an integer
                              # 'None' means full tree till single leaf
           )
ct.fit(x_train,y_train)
```

```python
y_te = ct.predict(x_test)
```

```python
accuracy_score(y_test,y_te)
```

# Decision Tree Classifier as Gini

```python
from sklearn.tree import DecisionTreeClassifier as dtree
ct = dtree(
           criterion="gini",   
           max_depth=None       # Alternative, specify an integer
                              # 'None' means full tree till single leaf
           )
ct.fit(x_train,y_train)
```

```python
y_tes = ct.predict(x_test)
```

```python
accuracy_score(y_test,y_tes)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=100)
clf.fit(x_train,y_train)

y_pred1=clf.predict(x_test)
accuracy_score(y_test,y_pred1)
```
